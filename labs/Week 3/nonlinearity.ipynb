```python
import pandas
import matplotlib.pyplot
import sklearn.linear_model
import time
```

```python
df = pandas.read_csv('SYD_DAT_5/data/anscombe.csv', sep='\t')
```

```python
clf = sklearn.linear_model.LinearRegression()
clf.fit(df.loc[:,['one_x']], df.one_y)
theil = sklearn.linear_model.TheilSenRegressor()
theil.fit(df.loc[:,['one_x']], df.one_y)
ransac = sklearn.linear_model.RANSACRegressor()
ransac.fit(df.loc[:,['one_x']], df.one_y)
```

```python
ransac.predict(df.loc[:,['one_x']])
```

```python
ransac.estimator_
```
