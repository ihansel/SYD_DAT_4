```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
column_names = ['sepal_length_in_cm','sepal_width_in_cm','petal_length_in_cm','petal_width_in_cm','species']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=column_names)
setosa = iris[iris.species == 'Iris-setosa']
versicolor = iris[iris.species == 'Iris-versicolor']
virginica = iris[iris.species == 'Iris-virginica']
```
