1) The article tackles the recent phenomena that is “Data Science”; specifically, offering a view on the type of “Data Scientist” currently in the work force. This article offers a view on 4 “main” types of “Data Scientists”.

It is important to note that different types of “Data Scientists” will posses these skills in different quantities and to varying degrees of expertise (e.g. a Data Businessperson may possess more Business knowledge compared to a Data Developer but will only have a limited knowledge in programming compared to a Data Developer).

	♣	Data Businesspeople: Focused more on the business side of an organisation and most likely to be managing people and teams. Data Businesspeople, in my opinion, will most likely be elected to manage projects or teams in this field. 	♣	Data Researchers: More theoretical and more likely to have stemmed their careers from a academic position, Data Researchers can provide an organisation with knowledge and a deep understanding of complex data concepts. ## Exercise Four - Movie Lens Data

```python
# for each occupation in 'users', count the number of occurrences
import pandas as pd
pd.read_table('u.user')
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = pd.read_table('u.user', sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})
users.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
byoccupation = users.groupby('occupation')
byoccupation['age'].mean()
```

```python
# for each occupation, calculate the minimum and maximum ages
byoccupation['age'].agg([min,max])
```

```python
# for each combination of occupation and gender, calculate the mean age
byoccupation_gender = users.groupby(['occupation','gender'])
byoccupation_gender['age'].mean()
```

```python
# randomly sample a DataFrame
users.sample(n=10)
```
