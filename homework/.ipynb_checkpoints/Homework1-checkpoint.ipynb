```python
# read 'u.user' into 'users'
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
users = df.read_table('u.user', sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})
```

```python
users
```

```python
# for each occupation in 'users', count the number of occurrences

users.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
#users.age.mean() 
users.groupby('occupation').age.mean()
```

```python
# for each occupation, calculate the minimum and maximum ages

users.groupby('occupation').age.min()
users.groupby('occupation').age.max()
```

```python
# for each combination of occupation and gender, calculate the mean age

users.groupby(['occupation', 'gender']).age.mean()
```

# The Data ascience Handbook Summary#



## Chapter 25 From Undergraduate to Data Science

In this chapter the interviewer tells his history about his transition from university to a full time job in data science, making emphasis four main points of data science:

    1. Statistics
    2. Programming 
    3. Communication
    4. Domain knowledge and experience
    

He did mention that at the beginning of intership he did realize that being only an statistician with not programming skills will not be enough to achieve his goals as it will be a limitation in the number of thing that he could do; therefore, worked hard in order to get those skills.

He also understood that communication was a main part of being data scientis as he would need to tell the history and started helping as a teaching fellow which helped hime to communicate better his ideas while helping other students.

He also highlighted that having a good understanding about the maths and statistics helps to articulate better at the time of explainingthe results.

In summary to be a data scientist is not enough just being good in one of the four main points mentioned above, you need all of them in order to have better intuition at the time to analyze the data a suggest better options at the time of chossing a solution for the business

## Chapter 13 The Interplay Between Science, Engineering and Data Science

In this chapter Diane Wu did talk about her transition first from computer science to bioinformatics and then from their PHD in Computational Genetics to data science.

She did mention that the transition from the PHD academia to the data science role was not too complicated as she will keep using same tools, methods and algoritms but in a different way.

She also explains that her company vision is to pick very complex from its customers and provide value add solutions based on the data provided; in order to do that, the data science team will try to scope the problems, narrow them down and work with the engineers in order to achieve the best solution.

One of the challenges that she have found is that some times the customer do not know what they want, and some times they want to be poved that data science is worth the investment.

Her advice for people who want to become a data science is not to be afraid or get overwhelmed with the requirenments; start picking up the needed skills instead one at the time; however, she point it out that is also important to understand what type of job you want "and realizing that you can’t be qualified for everything. You have to pick your best shot and hone your skills there."

## Analyzing the Analyzers 

1. There are four types of data scientist:
    - Data businesspeople.
    - Data creatives.
    - Data developers.
    - Data researchers.
    
 
2. The success of a data scientist lies in his/her ability to at least prototype every step needed to derive new insights.

3. A data scientist should have a substantial deep expertise in at least one of the following skills: 

    - Statistics.
    - Big data.
    - Business Communication.
    
 
4. The "T" shaped professionals have an advantage because they can work better in interdisciplinary projects.

5. With a better understanding of the roles and skills of the different types (clusters) of data scientists, it will be easier and more efficient recruitment process as well as much better undderstanding of what skills each data scientist needs to improve in order to achieve their goals and get their desired job
