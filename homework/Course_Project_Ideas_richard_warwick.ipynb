## Idea 3: Digital Media - Accounting for Cookie Bombing

### Overview

There is a phenomenon in digital marketing colloquially referred to as 'cookie bombing' where vast numbers of low cost advertising inventory is used to place cookies on as many computers as possible so that if some kind of desired outcome is achieved, such as signing up for a credit card, the provider of that inventory can claim credit for the acquisition, and will be automatically favoured for more investment or future spend.

The issue with this is that a lot of inventory cheap inventory is not even viewable, so there are occurences where an ad is recieving credit even though there is no way that the user saw it.

The challenge is can I use data science methods to identify and approximate the impact of 'cookie bombing' and develop a model for attribution that takes into account its probability

### What Data I Will Use

Anonymised 1st party data from a digital marketing platform - ddm.google.com/bidmanager/
10m-50m samples, 40+ dimensions
100 GB csvs could do api

### What is the Outcome I am Trying to Achieve?

More accurate attribution modelling for digital advertising campaigns

### What modeling techniques might it involve?

Not sure
