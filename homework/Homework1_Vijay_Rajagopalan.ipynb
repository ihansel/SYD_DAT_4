## Exercise Four - Movie Lens Data

```python
import pandas as pd
```

```python
# for each occupation in 'users', count the number of occurrences
data_columns = ["user_id", "age", "gender", "occupation", "zip_code"]
data = pd.read_table('../labs/Week 1/u.user', sep="|", header=None, names=data_columns)
data.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
data.groupby('occupation').age.mean()
```

```python
# for each occupation, calculate the minimum and maximum ages

df_occupation = data.groupby('occupation')
df_age_min = df_occupation.age.min()
df_age_max = df_occupation.age.max()
df_age_min_max = pd.concat([df_age_min, df_age_max], axis = 1)
df_age_min_max.columns = ['Age (Min)', 'Age (Max)']
df_age_min_max
```

```python
# for each combination of occupation and gender, calculate the mean age
data.groupby(['occupation', 'gender']).age.mean()
```

```python
# randomly sample a DataFrame
# sample of 10 items
data.sample(10)
```

```python
# detect duplicate users
len(pd.unique(data.user_id)) == data.user_id.count()
```

Length of Unique User IDs and Total Count of User IDs is equal, so atleast User IDs are all distinct. 

So, to find duplicate users, we will have to search the other columns. The probability of having the same age, gender and occupation is likely but also having the same zip_code is very unlikely. So we can group the Data Frame by zip_code and then find same values in other columns.

```python
df_zip_code = data.groupby('zip_code')
df1 = df_zip_code.user_id.count() > 1
df2 = df1[df1 == True]
#data[data.zip_code.isin(df2.axes)]
#non_unique_zip_code.axes
#df_zip_code[non_unique_zip_code.tolist()]
#type(non_unique_zip_code.axes)
#data[data.zip_code.isin(non_unique_zip_code.axes)]
```
