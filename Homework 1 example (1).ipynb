```python
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
```

```python
users = pd.read_table('u.user', sep='|', header=None, names=user_cols, index_cols='user_id', dtype={'zip_code':str})
```

```python
users = pd.read_table('u.user', sep='|', header=None, names=user_cols, index_col='user_id', dtype={'zip_code':str})
```

```python
users
```

```python
# for each occupation in 'users', count the number of occurrences
users.occupation.value_counts()
```

```python
# for each occupation, calculate the mean age
byoccupation = users.groupby('occupation')
byoccupation['age'].mean()
```

```python
# for each occupation, calculate the minimum and maximum ages
byoccupation = users.groupby('occupation')
byoccupation['age'].max()
```

```python
byoccupation = users.groupby('occupation')
byoccupation['age'].min()
```

```python
# for each combination of occupation and gender, calculate the mean age
bygender = users.groupby(['occupation','gender'])
bygender['age'].mean()
```

```python
# randomly sample a DataFrame
import random
rdm = users.groupby(['occupation','gender','age', 'zip_code'])
rows = random.sample(rdm, 10)
rows
```
